```python
import os
import folium
import pandas as pd 

print(folium.__version__)
```

```python
import numpy as np

#data = (np.random.normal(size=(100, 3)) *
       # np.array([[1, 1, 1]]) +
       # np.array([[48, 5, 1]])).tolist()
df = pd.read_csv('citylocandvalue.csv')
```

```python
from folium.plugins import HeatMap

max_amount = float(df['Value'].max())

m = folium.Map([56,-95], tiles='Stamen Terrain', zoom_start=4)

hm_wide = HeatMap(list(zip(df.Lat.values, df.Lon.values, df.Value.values)), 
                   min_opacity=0.2,
                   max_val=max_amount,
                   radius=17, blur=15, 
                   max_zoom=1, 
                 )


m.add_child(hm_wide)

m.save(os.path.join('results', 'Heatmap.html'))

m
```
